# CURVAS DE NIVEL PARA UNA FUNCIÓN DEFINIDA EN $\mathbb{R} ^3$  

En este caso se busca mostrar las curvas de nivel (una gráfica tipo *contour*) para la función definida como sigue:

$f: \mathbb{R} ^3 \rightarrow \mathbb{R} ^3$ mediante $(x,y,z) \rightarrow (x,y,x + y^2)$

Vamos a llevar a cabo este ejercicio para el mapeo en el intervalo $[-10,10]$.

```python
#Importamos la paquetería para incluir las matemáticas que no están por defecto en Python, así como la paquetería para graficar:
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Usamos la función np.meshgrid() para llamar al gráfico tipo contour y np.linspace() para definir el intervalo:
x, y = np.meshgrid(np.linspace(-10,10,201),np.linspace(-10,10,201))
#Por cuestiones del suavizado en la visualización de datos, tomamos la convención de que el número de valores entre -10 y 10 
#sea de 201.
z = x + y**2
```

```python
curvas = plt.contour(x,y,z) #Esta línea grafica las curvas de nivel de f
plt.clabel(curvas, inline = 1, fontsize = 10) #Esta línea imprime los valores en la gráfica
plt.title(r'$f(x,y,x)=(x,y,x+ y^2)$', fontsize=16)
```

Podemos también llenar el área, como una especie de mapa de calor:

```python
plt.contourf(x,y,z)
plt.colorbar() #Muestra la barra indicadora de los valores asociados 
```

Finalmente, como extensión de lo anterior, podemos visualizar la gráfica tridimensional de la función definida:

```python
Graf3D=plt.axes(projection="3d")
Graf3D.contour3D(x,y,z,1000,cmap='viridis')
```

La gráfica tridimensional puede ser vista desde arriba para visualizar el mapa de calor anterior:

```python
Graf3D=plt.axes(projection="3d")
Graf3D.contour3D(x,y,z,1000,cmap='viridis')
Graf3D.view_init(70,30)
```
