La onda sera difractada por el borde y la intensidad resultante en la posición $ (x,y) $ marcada por el punto esta dada por la teoría de difracción de campo cercano:

$I=\frac{I_0}{8} \cdot ([2C(u)+1]^2+[2S(u)+1]^2)$

Donde $I_0$ es la intensidad de la función de onda antes de la difracción, mientras que:

$u=x \sqrt{\frac{2}{\lambda z}}$ , $C(u)=\int\limits_0^u cos(\frac{1}{2}\pi t^2)dt$ , $S(u)=\int\limits_0^u sen(\frac{1}{2}\pi t^2)dt$

Se busca graficar la intensidad resultante $I$ para tres distintas $I_0$ en un intervalo $x = [−5, 5]$
metros; supón el caso de una onda de sonido ($\lambda$ = 1, y $z$ = 3m). Además, calcula las integrales
con el método de Simpson.

```python
from math import *
from pylab import *
import numpy as np

def simpson(f,a,b,N):
    h=(b-a)/N
    s=0.0
    x=a+h
    for i in range(1,N,2):
        s+=4*f(x)
        x+=2*h
    
    x=a+2*h
    for i in range(2,N,2):
        s+=2*f(x)
        x+=2*h
    return (h/3)*(f(a)+f(b)+s)
```

```python
#I = (I0/8)*((2*C+1)**2 + (2*S+1)**2)
#u = x*((2/l*z)**(1/2))
C = lambda t:np.cos((np.pi*t**2)/2)
S = lambda t:np.sin((np.pi*t**2)/2)

lista_u=[]
lista_x=[]
a,b=-5,5
N=1000
h=(b-a)/N
for x in arange(a,b,h):
    lista_x.append(x)
    l = 1
    z = 3 #metros
    u = x*((2/l*z)**(1/2))
    lista_u.append(u)

lista_C=[]
for i in lista_u:
    c = simpson(C,0,i,1000)
    lista_C.append(c)
    
lista_S=[]
for i in lista_u:
    s = simpson(S,0,i,1000)
    lista_S.append(s)
```

Gráfica 1:

```python
lista_I1=[]
I0 = 1
for i in range(0,len(lista_C)):
    I = (I0/8)*((2*lista_C[i]+1)**2 + (2*lista_S[i]+1)**2)
    lista_I1.append(I)
    
plot(lista_x,lista_I1)
plt.xlabel("x",size=16)
plt.ylabel("I(x)",size=16)
plt.grid()
plt.title("Intensidad de onda resultante. I_0 = 1",size=20)
plt.show()
```
