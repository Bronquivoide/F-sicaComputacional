# INTERPOLACIÓN:

Haz tu propia función llamada $Lagrange(n,x,y)$, que reciba los nodos o puntos para
interpolar $x$,$y$ y el grado del polinomio deseado $n$.

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

print("Este programa aplica el método de interpolación de Lagrange para aproximar una serie de puntos a un polinomio. \n")
p_x=[]
p_y=[]
x = float(input("¿Que valor desea aproximar? \n"))
Nodos = int(input("¿Cuantos puntos tienes para interpolar? \n"))

for i in range(Nodos):
    A=float(input("Introduce la coordenada x del punto P(%d): \n" %(i)))
    p_x.append(A)
    B=float(input("Introduce la coordenada y del punto P(%d): \n" %(i)))
    p_y.append(B)

xi=np.array(p_x)
yi=np.array(p_y)

def lagrange(x,xi,yi):
    n = len(xi)
    polinomio = 0
    for i in range(0,n,1):
        numerador = 1
        denominador = 1
        for j  in range(0,n,1):
            if (j!=i):
                numerador = numerador*(x-xi[j])
                denominador = denominador*(xi[i]-xi[j])
                terminoLi = numerador/denominador
                
        polinomio = polinomio + terminoLi*yi[i]
        

    return polinomio

# Puntos gráfica
a = np.min(xi)
b = np.max(xi)
pxi = np.linspace(a,b)

res = lagrange(x,xi,yi)
print("\n")
print("El resultado de la aproximación para el valor:",x,"es:",res)


# Gráfica
plt.plot(xi,yi,'o', label = 'Puntos')
plt.plot(pxi,lagrange(pxi,xi,yi), label = 'Polinomio')
plt.legend()
plt.title('Interpolación Lagrange')
plt.show()
```
