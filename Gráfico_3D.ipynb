# CURVA EN  $\mathbb{R} ^3$: LA HÉLICE

En coordenadas rectangulares (o cartesianas, $(x,y,z)$), definimos la curva *hélice* como aquella que presenta la siguiente parametrización:

$x(t)=cos(t)$

$y(t)=sen(t)$

$z(t)=t$


El objetivo de este notebook es visualizar un gráfico en $\mathbb{R} ^3$ de la función anterior, descrita como:

$h: \mathbb{R}  \rightarrow \mathbb{R} ^3 $, mediante  $ t\rightarrow (cos(t),sen(t),t)$

En este caso vamos a trabajar mediante la siguiente notación de parametrización:

$z \rightarrow (x=cos(z), y=sen(z), z)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0,30,301)
x1 = np.sin(z1)
y1 = 2*np.cos(z1)
plt3D.plot3D(x1, y1, z1, color='g')
```

Jugando un poco con la gráfica:

```python
#Podemos definirla en un intervalo x0 <= x <=x1, y0 <= y <=y1 por ejemplo, -2 <= x,y <= 2
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0,30,301)
x1 = np.sin(z1)
y1 = 2*np.cos(z1)
plt3D.plot3D(x1, y1, z1, color='r')
plt.xlim([-2,2])
plt.ylim([-2,2])
```

```python
#Podemos definirla hacia otra dirección:
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0,30,301)
x1 = np.sin(z1)
y1 = 2*np.cos(z1)
plt3D.plot3D(z1, y1, x1, color='b')
```
